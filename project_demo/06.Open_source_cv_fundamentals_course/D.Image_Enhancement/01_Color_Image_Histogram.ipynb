```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('Road_test.jpg', 1)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```

# plot Plot histogram

```python
plt.hist(img.ravel(), 256)   #ravel()   256 2D to 1D Grouping of 256 gray levels
plt.show()
```

# plotopencvplot plots the histogram values returned by opencv

```python
histb = cv2.calcHist([img], [1], None, [256], [0, 255])
plt.plot(histb, color='b')
plt.show()
```
